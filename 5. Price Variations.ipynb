```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
monthly_data = pd.read_csv('Deseasonalized_Mandi_Prices.csv')
msp_data = pd.read_csv('Outlier_removed_msp_data.csv')
```

### Yearly variations


```python
var_2014 = monthly_data[monthly_data['Year']==2014]
var_2015 = monthly_data[monthly_data['Year']==2015]
var_2015.reset_index(inplace=True,drop=True)
var_2016 = monthly_data[monthly_data['Year']==2016]
var_2016.reset_index(inplace=True,drop=True)
```

```python
data_2014 = var_2014.groupby(['APMC','commodity']).size().reset_index()
data_2014.drop(columns=[0],inplace=True)
for i,row in data_2014.iterrows():
    temp_df = var_2014[(var_2014['APMC']==row.APMC)&(var_2014['commodity']==row.commodity)]
    if((len(temp_df.max_price.values != 0))& (len(temp_df.max_price.values !=0))):
        fluctuation = max(temp_df.max_price.values) - min(temp_df.min_price.values)
        data_2014.loc[i,'Fluctuation'] = fluctuation
    else:
        data_2014.loc[i,'Fluctuation'] = 0
data_2014 = data_2014.sort_values(by = 'Fluctuation',ascending=False)
data_2014 = data_2014[:20]    # Taking only top 20 rows of highest fluctuations
data_2014.reset_index(inplace=True,drop=True)
data_2014
```
